## Part 2: Transfer learning

As established in Part 1, our dataset is simply too small to stably train a deep network from scratch. To overcome this limitation, we will employ Transfer Learning. Instead of starting with randomized weights, we will first train our custom DeepCNN on a large proxy dataset (CIFAR-10). This gives the model a massive "head start" by teaching it how to extract universal visual features. Like edges, gradients, and basic shapesâ€”from a diverse set of images (10 classes 6000 per class). Once pre-trained, we will transfer those learned weights and fine tune the network on our specific bathroom dataset.

First, the setup. Identical to Part 1.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np
import time
import copy
import os
import torch.nn.functional as F

# Configuration
DATA_PATH = 'data' 
BATCH_SIZE = 32
IMG_SIZE = 300    
VALID_SPLIT = 0.2
SEED = 42
NORM_MEAN = [0.485, 0.456, 0.406]
NORM_STD = [0.229, 0.224, 0.225]
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
# Transforms for dataset
data_transforms = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(NORM_MEAN, NORM_STD)
])

# Load dataset
try:
    full_dataset = datasets.ImageFolder(root=DATA_PATH, transform=data_transforms)
    class_names = full_dataset.classes
    
    # Split
    val_size = int(len(full_dataset) * VALID_SPLIT)
    train_size = len(full_dataset) - val_size
    generator = torch.Generator().manual_seed(SEED)
    train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size], generator=generator)

    # Loaders
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)
    
    print(f"Custom Data Loaded: {len(train_dataset)} training images.")
except:
    print("Warning: Could not load custom data yet. Make sure 'data' folder exists.")
```

### Base model from Part 1 (after all the improvements).

```python
class DeepCNN(nn.Module):
    def __init__(self, num_classes=3):
        super(DeepCNN, self).__init__()
        # Block 1
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.pool = nn.MaxPool2d(2, 2)
        # Block 2
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        # Block 3
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(128)
        # Block 4
        self.conv4 = nn.Conv2d(128, 256, kernel_size=3, padding=1)
        self.bn4 = nn.BatchNorm2d(256)
        
        self.gap = nn.AdaptiveAvgPool2d((1, 1))
        self.dropout = nn.Dropout(0.5)
        self.fc = nn.Linear(256, num_classes)

    def forward(self, x):
        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.pool(F.relu(self.bn2(self.conv2(x))))
        x = self.pool(F.relu(self.bn3(self.conv3(x))))
        x = self.pool(F.relu(self.bn4(self.conv4(x))))
        x = self.gap(x)
        x = x.flatten(1)
        x = self.dropout(x)
        x = self.fc(x)
        return x
```

```python
import torchvision
from torchvision import datasets
import torch.nn.functional as F

# 1. Configuration for CIFAR
print(f"Goal: Pre-train DeepCNN on 60,000 CIFAR images to learn features.")

# CIFAR-10 stats
CIFAR_MEAN = [0.4914, 0.4822, 0.4465]
CIFAR_STD = [0.2023, 0.1994, 0.2010]

# Simple transforms for CIFAR (32x32 is native size)
cifar_transforms = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(CIFAR_MEAN, CIFAR_STD)
])
```

Download and train CIFAR10:

```python
# 2. Download and Load CIFAR-10
print("Downloading CIFAR-10...")
# We use a batch size of 64 for faster training on the larger dataset
cifar_train = datasets.CIFAR10(root='./data_cifar', train=True, download=True, transform=cifar_transforms)
cifar_loader = DataLoader(cifar_train, batch_size=64, shuffle=True, num_workers=2)

# 3. Define the Model (Output = 10 classes)
class DeepCNN_CIFAR(nn.Module):
    def __init__(self):
        super(DeepCNN_CIFAR, self).__init__()
        # EXACT SAME BODY as our Part 1 model
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.pool = nn.MaxPool2d(2, 2)
        
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(128)
        
        self.conv4 = nn.Conv2d(128, 256, kernel_size=3, padding=1)
        self.bn4 = nn.BatchNorm2d(256)
        
        self.gap = nn.AdaptiveAvgPool2d((1, 1))
        self.dropout = nn.Dropout(0.5)
        
        # 10 classes for CIFAR (Plane, Car, Bird, etc.)
        self.fc = nn.Linear(256, 10) 

    def forward(self, x):
        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.pool(F.relu(self.bn2(self.conv2(x))))
        x = self.pool(F.relu(self.bn3(self.conv3(x))))
        x = self.pool(F.relu(self.bn4(self.conv4(x))))
        x = self.gap(x)
        x = x.flatten(1)
        x = self.dropout(x)
        x = self.fc(x)
        return x

# 4. Train on CIFAR (The "Pre-training")
# We train for 10 epochs. =
print("\nTraining DeepCNN on CIFAR-10 (10 Epochs)...")
model_cifar = DeepCNN_CIFAR().to(device)
optimizer_cifar = optim.Adam(model_cifar.parameters(), lr=0.001)
criterion_cifar = nn.CrossEntropyLoss()

model_cifar.train()
for epoch in range(10): 
    running_loss = 0.0
    for inputs, labels in cifar_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer_cifar.zero_grad()
        outputs = model_cifar(inputs)
        loss = criterion_cifar(outputs, labels)
        loss.backward()
        optimizer_cifar.step()
        running_loss += loss.item()
    print(f"Epoch {epoch+1}/10 - Loss: {running_loss/len(cifar_loader):.4f}")

# 5. Save the CIFAR Weights
torch.save(model_cifar.state_dict(), 'deep_cnn_cifar_pretrained.pth')
print("Pre-trained CIFAR model saved!")
```

To implement transfer learning, we first need a bigger dataset to act as a "computer vision school" for our network. We selected CIFAR-10, a well-known public dataset containing 60,000 diverse, colorful images across 10 distinct classes (like birds, cars, and airplanes).

* Why this dataset is relevant: Even though CIFAR-10 does not contain images of bathrooms, the fundamental rules of processing visual data are universal. Training on a large, diverse dataset prevents the model from memorizing specific pixels and forces it to learn robust, generalizable computer vision skills.

* What features it teaches the model:  During this pre-training phase, the early convolutional layers learn to detect low-level universal features like edges, contrasting color boundaries, and gradients. The deeper layers learn to assemble these into textures, curves, and basic geometric shapes. The visual "building blocks" of a car or a dog are the exact same building blocks needed to recognize a toilet or a sink.

* Architecture Adjustments: We used the exact same DeepCNN architecture from our final Part 1 experiment (including Batch Normalization and Dropout). The only required adjustment was the final classification head: we changed nn.Linear(256, 3) to nn.Linear(256, 10) to match the 10 classes of the CIFAR dataset.

```python
# Same utility function from part 1
def train_model(model, criterion, optimizer, num_epochs=10):
    since = time.time()

    history = {
        'train_loss': [], 'train_acc': [],
        'val_loss': [], 'val_acc': []
    }

    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0

    for epoch in range(num_epochs):
        print(f'Epoch {epoch+1}/{num_epochs}')
        print('-' * 10)

        # Each epoch has a training and validation phase
        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()  # Set model to training mode
                dataloader = train_loader
            else:
                model.eval()   # Set model to evaluate mode
                dataloader = val_loader

            running_loss = 0.0
            running_corrects = 0

            # Iterate over data.
            for inputs, labels in dataloader:
                inputs = inputs.to(device)
                labels = labels.to(device)

                # Zero the parameter gradients
                optimizer.zero_grad()

                # Forward
                # Track history if only in train
                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    # Backward + optimize only if in training phase
                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                # Statistics
                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

            epoch_loss = running_loss / len(dataloader.dataset)
            epoch_acc = running_corrects.double() / len(dataloader.dataset)

            print(f'{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

            # Save history
            if phase == 'train':
                history['train_loss'].append(epoch_loss)
                history['train_acc'].append(epoch_acc.item())
            else:
                history['val_loss'].append(epoch_loss)
                history['val_acc'].append(epoch_acc.item())

                # Deep copy the model if it's the best one so far
                if epoch_acc > best_acc:
                    best_acc = epoch_acc
                    best_model_wts = copy.deepcopy(model.state_dict())

    time_elapsed = time.time() - since
    print(f'\nTraining complete in {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s')
    print(f'Best Val Acc: {best_acc:.4f}')

    # Load best model weights
    model.load_state_dict(best_model_wts)
    return model, history
```

Now that our model has a solid foundation in feature extraction from the CIFAR-10 pre-training, we adapt it to our specific task through a process called fine-tuning:
* We iterate through the convolutional body and set requires_grad = False. This "freezes" the weights.
* The old classification head was designed to output 10 CIFAR classes. We discard it and attach a fresh, untrained linear layer designed for our 3 specific bathroom classes.
* We configure our Adam optimizer to only look at parameters that require gradients.

```python
# 1. Define the architecture (Must match the saved CIFAR model structure initially)
class DeepCNN_Transfer(nn.Module):
    def __init__(self):
        super(DeepCNN_Transfer, self).__init__()
        # EXACT SAME BODY
        # Block 1
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.pool = nn.MaxPool2d(2, 2)
        
        # Block 2
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        
        # Block 3
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(128)
        
        # Block 4
        self.conv4 = nn.Conv2d(128, 256, kernel_size=3, padding=1)
        self.bn4 = nn.BatchNorm2d(256)
        
        # Head
        self.gap = nn.AdaptiveAvgPool2d((1, 1))
        self.dropout = nn.Dropout(0.5)  # Drop 50% of neurons
        
        # INITIAL HEAD: 10 classes
        self.fc = nn.Linear(256, 10) 

    def forward(self, x):
        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.pool(F.relu(self.bn2(self.conv2(x))))
        x = self.pool(F.relu(self.bn3(self.conv3(x))))
        x = self.pool(F.relu(self.bn4(self.conv4(x))))
        x = self.gap(x)
        x = x.flatten(1)
        x = self.dropout(x)
        x = self.fc(x)
        return x

# 2. Instantiate and Load Weights
print("Loading Pre-trained CIFAR 'Brain'...")
model_transfer = DeepCNN_Transfer().to(device)
model_transfer.load_state_dict(torch.load('deep_cnn_cifar_pretrained.pth'))
print("Weights Loaded!")

# 3. FREEZE the Body
# We turn off gradient calculation for all existing parameters
for param in model_transfer.parameters():
    param.requires_grad = False

print("Convolutional Layers Frozen (We won't retrain them).")

# 4. Replace the Head
# Since we are assigning a NEW layer, its 'requires_grad' will default to True (Trainable)
num_ftrs = model_transfer.fc.in_features  # Should be 256
model_transfer.fc = nn.Linear(num_ftrs, 3) # NEW Head: our 3 Classes
model_transfer = model_transfer.to(device)

print("New Head Attached (Output: 3 Classes). Ready for Fine-Tuning.")

# 5. Fine-Tune on our Dataset
# Note: We only optimize the parameters that require gradients (the new head)
optimizer_ft = optim.Adam(filter(lambda p: p.requires_grad, model_transfer.parameters()), lr=0.001)
criterion = nn.CrossEntropyLoss()

print("\nStarting Transfer Learning (Fine-Tuning)...")
# We use our original train_loader (from the bathroom dataset)
# Training for 15 epochs should be plenty since the body is already smart
model_transfer, hist_transfer = train_model(model_transfer, criterion, optimizer_ft, num_epochs=15)

torch.save(model_transfer.state_dict(), 'deep_cnn_transfer_learning.pth')
```

```python
def plot_transfer_results(history, title, baseline=0.63):
    """
    Function to plot training history against a baseline.
    Args:
        history: The history dictionary from train_model (contains 'train_acc', 'val_acc', etc.)
        title: String to add to the graph title
        baseline: The accuracy of your best Part 1 model
    """
    plt.figure(figsize=(12, 5))

    # --- Plot Accuracy ---
    plt.subplot(1, 2, 1)
    plt.plot(history['train_acc'], label='Train Acc')
    plt.plot(history['val_acc'], label='Val Acc')
    # Add the Red Baseline Line
    plt.axhline(y=baseline, color='r', linestyle='--', label=f'Baseline (Scratch): {baseline}')
    plt.title(f'Accuracy: {title}')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # --- Plot Loss ---
    plt.subplot(1, 2, 2)
    plt.plot(history['train_loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Val Loss')
    plt.title(f'Loss: {title}')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()
```

As the training graphs illustrate, our initial transfer learning attempt underperformed the best baseline model trained from scratch. Both training and validation accuracies peaked at roughly 0.56, falling short of the 0.63 baseline marked by the red dashed line. However, this drop in performance came with a clear computational trade-off: because we only trained the new classification head while keeping the base feature extractor completely frozen, this run took only about 10 minutes to train, compared to the ~15 minutes required for the baseline model. While the loss curves show a steady, non-overfitting decline, the frozen base clearly lacked the flexibility needed to capture the specific patterns in our dataset.  
One fact that I initially forgot that can be the reason for it is that CIFAR10 images are all 32x32 pixels, while our images are 300 by 300. Therefore the next attempt became the unfreezing of the last convolution layers to let the model readjust to a different resolution.

```python
import torch.optim as optim

print("Unfreezing the last convolutional block for Fine-Tuning...")

# 1. Unfreeze the last block (Conv4 + BN4)
# We want the model to adjust its high-level features to the new image size
for param in model_transfer.conv4.parameters():
    param.requires_grad = True
for param in model_transfer.bn4.parameters():
    param.requires_grad = True

# 2. Lower the Learning Rate significantly (CRITICAL STEP)
# We use 1e-4 (10x smaller than before) because we don't want to
# destroy the delicate pre-trained weights; we just want to nudge them.
optimizer_finetune = optim.Adam(filter(lambda p: p.requires_grad, model_transfer.parameters()), lr=1e-4)

print("Restarting Training (Fine-Tuning Last Block)...")

# 3. Train for another 15 epochs
# Note: We pass the SAME model instance, so it starts from ~55% accuracy, not 0%
model_transfer, hist_finetune = train_model(model_transfer, criterion, optimizer_finetune, num_epochs=15)

# 4. Save this 'Fine-Tuned' version
torch.save(model_transfer.state_dict(), 'deep_cnn_finetuned.pth')
print("Fine-Tuned Model Saved!")
```

```python
plot_transfer_results(hist_finetune, "Finetuning last layer")
```

* The accuracy and loss curves are very erratic, with massive spikes and drops rather than a smooth learning progression. This is a classic symptom of noisy gradient updates, which is heavily amplified when training a complex model on a small dataset of around 300 images. Even with the learning rate drastically reduced to 1e-4, the model struggles to settle into a stable minimum.
* Despite unfreezing the conv4 block to allow the network to adjust its high-level features, the model's validation accuracy only peaked at 62%.
* The suspected problem remains the same: resolution differences. Lets try to solve it differently.

```python
# --- CRITICAL CHANGE: Resize to 32x32 to match CIFAR ---
IMG_SIZE_CIFAR = 32 

print(f"Reloading Dataset resized to {IMG_SIZE_CIFAR}x{IMG_SIZE_CIFAR}...")

# Defines transforms with the NEW size
cifar_compatible_transforms = transforms.Compose([
    transforms.Resize((IMG_SIZE_CIFAR, IMG_SIZE_CIFAR)),
    transforms.ToTensor(),
    transforms.Normalize(CIFAR_MEAN, CIFAR_STD)
])

# Load the dataset with these new transforms
try:
    full_dataset_32 = datasets.ImageFolder(root=DATA_PATH, transform=cifar_compatible_transforms)
    
    # Split again
    val_size = int(len(full_dataset_32) * VALID_SPLIT)
    train_size = len(full_dataset_32) - val_size
    generator = torch.Generator().manual_seed(SEED)
    train_dataset_32, val_dataset_32 = random_split(full_dataset_32, [train_size, val_size], generator=generator)

    # New Loaders
    train_loader_32 = DataLoader(train_dataset_32, batch_size=32, shuffle=True, num_workers=0)
    val_loader_32 = DataLoader(val_dataset_32, batch_size=32, shuffle=False, num_workers=0)
    
    print(f"Data Reloaded! Images are now 32x32 pixels.")
except:
    print("Error loading data. Check path.")
```

```python
# 1. Load the Pre-trained CIFAR Model
print("Loading Pre-trained CIFAR Model...")
model_transfer_32 = DeepCNN_Transfer().to(device)
model_transfer_32.load_state_dict(torch.load('deep_cnn_cifar_pretrained.pth'))

# 2. Freeze the Body
for param in model_transfer_32.parameters():
    param.requires_grad = False

# 3. Replace Head (10 -> 3 classes)
num_ftrs = model_transfer_32.fc.in_features
model_transfer_32.fc = nn.Linear(num_ftrs, 3).to(device)

# 4. Train (using the NEW 32x32 loaders)
print("Starting Transfer Learning on 32x32 Images...")
criterion = nn.CrossEntropyLoss()
optimizer_ft = optim.Adam(model_transfer_32.fc.parameters(), lr=0.001)

# Pass the NEW loaders explicitly by temporarily swapping variables 
temp_train = train_loader
temp_val = val_loader
train_loader = train_loader_32
val_loader = val_loader_32

model_transfer_32, hist_transfer_32 = train_model(model_transfer_32, criterion, optimizer_ft, num_epochs=15)

# Swap back just in case
train_loader = temp_train
val_loader = temp_val
```

```python
# 1. Unfreeze EVERYTHING
# We are already loaded with 32x32 data and the model is in memory.
# We just unlock the layers.
for param in model_transfer_32.parameters():
    param.requires_grad = True

print("Unfreezing All Layers for 32x32 Fine-Tuning...")

# 2. Use a Low Learning Rate 
# We use 1e-4 so we don't destroy the CIFAR weights, just nudge them.
optimizer_full_32 = optim.Adam(model_transfer_32.parameters(), lr=1e-4)

# 3. Train for 15 more epochs
print("Starting Full Fine-Tuning on 32x32 images...")

# Swap loaders again to be safe
temp_train = train_loader
temp_val = val_loader
train_loader = train_loader_32
val_loader = val_loader_32

model_transfer_32, hist_transfer_32_full = train_model(model_transfer_32, criterion, optimizer_full_32, num_epochs=15)

# Swap back
train_loader = temp_train
val_loader = temp_val

# Save
torch.save(model_transfer_32.state_dict(), 'deep_cnn_32_finetuned.pth')
```

In our final attempt with the CIFAR-10 base model, we decided to unfreeze all layers, allowing the entire network to fine-tune on our 32x32 resized dataset. This approach finally allowed us to break past our custom CNN's 63% baseline, achieving a peak validation accuracy of 68%.

However, this came at a cost: significant overfitting. As the graphs demonstrate, there is a distinct and growing divergence between the training and validation metrics. While the training accuracy climbs aggressively toward 84%, the validation accuracy struggles to keep pace, plateauing around the 68% mark. Similarly, the validation loss stops improving at the same rate as the training loss.

This behavior indicates that our model, now fully unfrozen and possessing maximum learning capacity, has begun memorizing the training data rather than learning generalizable features. Given that our dataset only contains around 300 images, squeezing them into a tiny 32x32 resolution destroyed too much critical visual detail. The network simply didn't have enough clear data to learn robust shapes and instead started memorizing the specific pixelated noise of the training set.

To solve this resolution bottleneck and allow a model to properly "see" our data, our next logical step is to abandon the 32x32 CIFAR constraints and upgrade to a powerful architecture designed natively for high-resolution images: ResNet50.

```python
from torchvision import models

print("Reloading Data for ResNet50 (224x224)...")

# ResNet expects 224x224 input
resnet_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(NORM_MEAN, NORM_STD)
])

# Reload datasets
full_dataset_resnet = datasets.ImageFolder(root=DATA_PATH, transform=resnet_transforms)

# Split
val_size = int(len(full_dataset_resnet) * VALID_SPLIT)
train_size = len(full_dataset_resnet) - val_size
generator = torch.Generator().manual_seed(SEED)
train_dataset_resnet, val_dataset_resnet = random_split(full_dataset_resnet, [train_size, val_size], generator=generator)

# Loaders (using the standard variables now)
train_loader = DataLoader(train_dataset_resnet, batch_size=32, shuffle=True, num_workers=0)
val_loader = DataLoader(val_dataset_resnet, batch_size=32, shuffle=False, num_workers=0)

print(f"Data Ready: {len(train_dataset_resnet)} training images at 224x224 resolution.")
```

First lets only replace the head:

```python
print("Loading ResNet50 (Pre-trained on ImageNet)...")

# 1. Load the Model
model_resnet = models.resnet50(weights='IMAGENET1K_V1')

# 2. Freeze the Body
# We want to keep all the smart filters it learned from ImageNet
for param in model_resnet.parameters():
    param.requires_grad = False

# 3. Replace the Head
# ResNet's final layer is called 'fc' 
num_ftrs = model_resnet.fc.in_features
model_resnet.fc = nn.Linear(num_ftrs, 3) # 3 Classes: Toilet, No-Toilet, Not-Bathroom

model_resnet = model_resnet.to(device)

print("ResNet50 Loaded & Adapted. Body is frozen, Head is new.")
```

```python
print("Starting Training with ResNet50...")

# We only train the head (fc layer)
optimizer_resnet = optim.Adam(model_resnet.fc.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()

# Train for fewer epochs because it learns fast
model_resnet, hist_resnet = train_model(model_resnet, criterion, optimizer_resnet, num_epochs=15)

# Save
torch.save(model_resnet.state_dict(), 'resnet50_finetuned.pth')
```

```python
plot_transfer_results(hist_resnet, "Transfer Learning Pretrained")
```

The results of this architectural shift are huge. By merely training the new classification head, the model shattered our custom CNN's 63% baseline, reaching a peak validation accuracy of 81.82%. The training process was also highly efficient since backpropagation only had to update the weights of the single final layer.

However, thre is still a mild overfitting. Toward the later epochs, a distinct gap forms. The training accuracy climbs past 90% and training loss continues to plummet, while validation accuracy plateaus around 81% and validation loss flattens out. To bridge this gap and allow the network to specialize, the model requires slightly more flexibility. Therefore, our next logical step is deeper fine-tuning by unfreezing the final convolutional layers of the ResNet50 base.

First, only the last layer:

```python
# 1. Reload ResNet fresh
print("Reloading ResNet50 for Unfreezing Experiment...")
model_unfrozen = models.resnet50(weights='IMAGENET1K_V1')

# 2. FREEZE First, then UNFREEZE the last block
# We want to keep the "Basic Features" (lines/colors) locked,
# but let the "High Level Features" (shapes/objects) adapt to bathrooms.
for param in model_unfrozen.parameters():
    param.requires_grad = False  # Lock everything initially

# Unfreeze the last block
for param in model_unfrozen.layer4.parameters():
    param.requires_grad = True
    
# 3. Replace the Head (Output Layer)
# Note: This is new, so it is automatically "unfrozen" (trainable)
num_ftrs = model_unfrozen.fc.in_features
model_unfrozen.fc = nn.Linear(num_ftrs, 3)
model_unfrozen = model_unfrozen.to(device)

# 4. Optimizer with Low Learning Rate
# We use 1e-4 (small) because we are fine-tuning pre-trained weights.
# If we used 0.001, we might destroy the knowledge it already has.
optimizer_unfrozen = optim.Adam(filter(lambda p: p.requires_grad, model_unfrozen.parameters()), lr=1e-4)
criterion = nn.CrossEntropyLoss()

# 5. Train
print("Starting Fine-Tuning (Layer 4 Unfrozen)...")
model_unfrozen, hist_unfrozen = train_model(model_unfrozen, criterion, optimizer_unfrozen, num_epochs=10)

# Save
torch.save(model_unfrozen.state_dict(), 'resnet50_unfrozen.pth')
plot_transfer_results(hist_unfrozen, "Layer 4 Unfrozen")
```

Now let's try the last 2 layers.

```python
# 1. Reload ResNet fresh
print("Reloading ResNet50 for Unfreezing Experiment...")
model_unfrozen = models.resnet50(weights='IMAGENET1K_V1')

# 2. FREEZE First, then UNFREEZE the last blocks
for param in model_unfrozen.parameters():
    param.requires_grad = False  # Lock everything initially

# Unfreeze the last two huge blocks (Layer 4 and Layer 3)
for param in model_unfrozen.layer4.parameters():
    param.requires_grad = True
for param in model_unfrozen.layer3.parameters():
    param.requires_grad = True

# 3. Replace the Head (Output Layer)
num_ftrs = model_unfrozen.fc.in_features
model_unfrozen.fc = nn.Linear(num_ftrs, 3)
model_unfrozen = model_unfrozen.to(device)

# 4. Optimizer with Low Learning Rate
optimizer_unfrozen = optim.Adam(filter(lambda p: p.requires_grad, model_unfrozen.parameters()), lr=1e-4)
criterion = nn.CrossEntropyLoss()

# 5. Train
print("Starting Fine-Tuning (Layers 3,4 Unfrozen)...")
model_unfrozen, hist_unfrozen = train_model(model_unfrozen, criterion, optimizer_unfrozen, num_epochs=5)

# Save
torch.save(model_unfrozen.state_dict(), 'resnet50_unfrozen.pth')
plot_transfer_results(hist_unfrozen, "Layers 3,4 Unfrozen")
```

No matter if its 1 layer or 2 layers that we unfreeze, we have full overfitting now. I even lowered the amount of epochs to 5 and we can see how fast it converges to that state of full 1.00 overfitting.  
This, again, is simply a model memorizing its training data. By unfreezing these deeper layers, we unlocked millions of trainable parameters. Because the dataset only contains roughly 300 images, the model has an overwhelming amount of capacity relative to the available data. Instead of learning to recognize the generalized features of a toilet or a sink, the network takes the easiest path.  
To prevent this memorization and force the network to learn generalizable features, we need to make the training task harder. Our next step is to implement Data Augmentation. Applying random flips, rotations, and color shiftsâ€”to artificially multiply our dataset's variety and break this overfitting loop.

### Data augmentation

```python
# 1. Define Augmentation Transforms (The "New" Data)
# We add flips, rotations, and color shifts to make the task harder (and the model smarter).
augment_transforms = transforms.Compose([
    transforms.Resize((224, 224)),              # Always resize first
    transforms.RandomHorizontalFlip(p=0.5),     # Flip left-right (Mirror)
    transforms.RandomRotation(degrees=15),      # Rotate slightly (+/- 15 deg)
    transforms.ColorJitter(brightness=0.2, contrast=0.2), # Change lighting
    transforms.ToTensor(),
    transforms.Normalize(NORM_MEAN, NORM_STD)
])

print("Reloading Data with Augmentation...")

# 2. Reload Dataset with Augmentation
# Note: We ONLY apply augmentation to the Training set. Validation stays 'clean'.
train_dataset_aug = datasets.ImageFolder(root=DATA_PATH, transform=augment_transforms)
# We use the same split indices to ensure fair comparison
train_subset_aug = torch.utils.data.Subset(train_dataset_aug, train_dataset_resnet.indices)

# Validation set uses the original (non-augmented) transforms
val_subset_clean = val_dataset_resnet 

# New Loaders
train_loader_aug = DataLoader(train_subset_aug, batch_size=32, shuffle=True, num_workers=0)
val_loader_clean = DataLoader(val_subset_clean, batch_size=32, shuffle=False, num_workers=0)

# 3. Setup Model
# Let's assume Layer 4 Unfrozen was good. We'll use that config.
model_aug = models.resnet50(weights='IMAGENET1K_V1')

# Freeze Body
for param in model_aug.parameters():
    param.requires_grad = False
# Unfreeze Layer 4 
for param in model_aug.layer4.parameters():
    param.requires_grad = True

# Replace Head
model_aug.fc = nn.Linear(model_aug.fc.in_features, 3)
model_aug = model_aug.to(device)

# 4. Train with Augmentation
print("Starting Training with Data Augmentation...")
optimizer_aug = optim.Adam(filter(lambda p: p.requires_grad, model_aug.parameters()), lr=1e-4)

# We temporarily swap the global loaders to use the augmented ones
temp_train = train_loader
temp_val = val_loader
train_loader = train_loader_aug
val_loader = val_loader_clean

model_aug, hist_aug = train_model(model_aug, criterion, optimizer_aug, num_epochs=10)

# Swap back
train_loader = temp_train
val_loader = temp_val

# Save
torch.save(model_aug.state_dict(), 'resnet50_augmented.pth')
```

```python
plot_transfer_results(hist_aug, "Pretrained + Data Augmentation")
```

As the graphs illustrate, this strategy disrupted the model's ability to fully memorize the images, however it still does so to an extent which can be seen from a gap between the blue and the orange lines. Despite that, this is our best model so far accuracy wise.
